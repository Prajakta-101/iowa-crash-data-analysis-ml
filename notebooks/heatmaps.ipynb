```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("Vehicle_Crashes_in_Iowa_20250410.csv")


# Select relevant numerical columns for correlation
columns_of_interest = [
    'Number of Fatalities', 
    'Number of Injuries', 
    'Number of Major Injuries', 
    'Number of Minor Injuries', 
    'Number of Possible Injuries', 
    'Number of Unknown Injuries', 
    'Amount of Property Damage', 
    'Number of Vehicles Involved', 
    'Total Number of Occupants'
]

# Compute the correlation matrix
correlation_matrix = dataset[columns_of_interest].corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=0, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python


# One-hot encode categorical variables
data = pd.get_dummies(data, columns=['Crash Severity', 'Weather Conditions', 'Light Conditions', 'Roadway Surface', 'Roadway Contribution'])

# Select relevant numerical columns, including the one-hot encoded columns
columns_of_interest = [
    'Number of Fatalities',
    'Number of Injuries',
    'Number of Major Injuries',
    'Number of Minor Injuries',
    'Number of Possible Injuries',
    'Number of Unknown Injuries',
    'Amount of Property Damage',
    'Number of Vehicles Involved',
    'Total Number of Occupants'
] + [col for col in data.columns if 'Crash Severity_' in col or 'Weather Conditions_' in col or 'Light Conditions_' in col or 'Roadway Surface_' in col or 'Roadway Contribution' in col]

# Compute the correlation matrix
correlation_matrix = data[columns_of_interest].corr()

# Create the heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap with Additional Factors')
plt.show()
```

```python
# One-hot encode specified categorical variables
data = pd.get_dummies(data, columns=['Manner of Crash/Collision', 'Environmental Conditions'])

# Select relevant numerical and one-hot encoded columns
columns_of_interest = [
    'Number of Fatalities',
    'Number of Injuries',
    'Amount of Property Damage',
    'Number of Vehicles Involved',
    'Total Number of Occupants'
] + [col for col in data.columns if 'Manner of Crash/Collision_' in col or 'Environmental Conditions_' in col]

# Compute the correlation matrix
correlation_matrix = data[columns_of_interest].corr()

# Create the heatmap
plt.figure(figsize=(24, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap with Manner of Crash/Collision and Environmental Conditions')
plt.show()
```
